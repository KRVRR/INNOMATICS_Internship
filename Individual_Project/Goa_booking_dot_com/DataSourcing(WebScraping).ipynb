<h4><b>Step.no.5.</b></h4><h7><b>Writing some code to scrap the required data</b></h7><br><br>

```python
#URL = 'https://www.booking.com/searchresults.en-gb.html?aid=393655&label=msn-q3iUVj90_tv5P24rA9nQEw-80539320026282%3Atikwd-16659121801%3Aloc-90%3Aneo%3Amtp%3Adec%3Aqspg%20hostels%20website%20in%20hyderabad&sid=a4fbe487c91ce0d1683a65096d4b7198&tmpl=searchresults&ac_click_type=b&ac_position=0&checkin_month=1&checkin_monthday=10&checkin_year=2021&checkout_month=1&checkout_monthday=11&checkout_year=2021&city=-2097701&class_interval=1&dest_id=4127&dest_type=region&dtdisc=0&from_sf=1&group_adults=2&group_children=0&inac=0&index_postcard=0&label_click=undef&no_rooms=1&postcard=0&raw_dest_type=region&room1=A%2CA&sb_price_type=total&search_selected=1&shw_aparth=1&slp_r_match=0&src=searchresults&src_elem=sb&srpvid=2f8360080d83022a&ss=Goa%2C%20India&ss_all=0&ss_raw=Goa&ssb=empty&sshis=0&ssne=Hyderabad&ssne_untouched=Hyderabad&top_ufis=1&rows=25&offset=1'
Titles = []
Prices = []
TotalReviews = []
ReviewRates = []
ReviewMarks = []
Accomodations = []
AdditionalCharges = []
CancellationPolicies = []
Beds = []
Locations = []
RoomsAvailables = []

      
for i in range(25,500 ,25):
    u = 'https://www.booking.com/searchresults.en-gb.html?aid=375653&label=msn-LOcHQGfA16aswDjFb7_rcw-80539252256335%3Atikwd-80539278535451%3Aloc-90%3Aneo%3Amte%3Adec%3Aqsbooking.com&sid=a4fbe487c91ce0d1683a65096d4b7198&tmpl=searchresults&checkin_month=12&checkin_monthday=30&checkin_year=2020&checkout_month=1&checkout_monthday=1&checkout_year=2021&class_interval=1&dest_id=4127&dest_type=region&from_sf=1&group_adults=2&group_children=0&label_click=undef&nflt=tripTypeThemeFilter%3D7%3B&no_rooms=1&raw_dest_type=region&room1=A%2CA&sb_price_type=total&shw_aparth=1&slp_r_match=0&src=searchresults&src_elem=sb&srpvid=d6ea584d52bb0040&ss=Goa&ssb=empty&ssne=Goa&ssne_untouched=Goa&top_ufis=1&rows=25&offset={}'. format(i)      
    page = requests.get(u,headers=HEADERS)
    soup = BeautifulSoup(page.content, "lxml")
    for x in soup.find_all('div', attrs={'class' : 'sr_item_content sr_item_content_slider_wrapper'}):
        
        T = x.find('span', attrs = {'class' : 'sr-hotel__name'})
        if T is None:
            Titles.append(np.NaN)
        else:
            Titles.append(T.text)

        C = x.find('sup', attrs = {'class' : 'sr_room_reinforcement e2e-free-cancellation bui-u-inline'})
        if C  is None:
            CancellationPolicies.append(np.NaN)
        else:
            CancellationPolicies.append(C.text)

        A = x.find('div', attrs = {'class' : 'prd-taxes-and-fees-under-price prco-inline-block-maker-helper blockuid-'})
        if A is None:
            AdditionalCharges.append(np.NaN)
        else:
            AdditionalCharges.append(A.text)

        B = x.find('div', attrs = {'class' : 'c-beds-configuration'})
        if B is None:
            Beds.append(np.NaN)
        else:
            Beds.append(B.text)
        
        Ac = x.find('div', attrs = {'class' : 'bui-price-display__label prco-inline-block-maker-helper'})        
        if Ac is None:
            Accomodations.append(np.NaN)
        else:
            Accomodations.append(Ac.text)  
            
        P = x.find('div', attrs = {'class' : 'bui-price-display__value prco-inline-block-maker-helper'})        
        if P is None:
            Prices.append(np.NaN)
        else:
            Prices.append(P.text)  
        
        Tr = x.find('div', attrs = {'class' : 'bui-review-score__text'})     
        if Tr is None:
            TotalReviews.append(np.NaN)
        else:
            TotalReviews.append(Tr.text)  
            
        Rr = x.find('div', attrs = {'class' : 'bui-review-score__badge'})        
        if Rr is None:
            ReviewRates.append(np.NaN)
        else:
            ReviewRates.append(Rr.text)  
            
        Rm = x.find('div', attrs = {'class' : 'bui-review-score__title'})       
        if Rm is None:
            ReviewMarks.append(np.NaN)
        else:
            ReviewMarks.append(Rm.text) 
            
        L = x.find('a', attrs = {'class' : 'bui-link'})    
        if L is None:
            Locations.append(np.NaN)
        else:
            Locations.append(L.text)
            
        Ra = x.find('span', attrs = {'class' : 'only_x_left sr_rooms_left_wrap'})   
        if Ra is None:
            RoomsAvailables.append(np.NaN)
        else:
            RoomsAvailables.append(Ra.text)
```

<h4><b>Step.no.6.</b></h4><h7><b>Creating a DataFrame</b></h7><br><br>

```python
#Iterates over a big list of .mp3 links to get the metadata tags and save it to an Excel file. Results in this error.
# Hence we use the following code
a = {'Title' : Titles,
     'Location': Locations,
     'Price' : Prices, 
     'TotalReviews' : TotalReviews, 
     'ReviewRate' : ReviewRates,
     'ReviewMark' : ReviewMarks,
     'Accomodation' : Accomodations,
     'RoomsAvailable' : RoomsAvailables,
     'Bed' : Beds,
     'AdditionalCharge' : AdditionalCharges,
     'CancellationPolicy' : CancellationPolicies}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
```
