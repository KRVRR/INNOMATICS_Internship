<h3>Question No.4. Find the correlation coefficient between the complete Study Percentage i.e Average of all the academics and the Salary ?</h3>

```python
df.insert(2,"Study_Percentage"," ")
```

```python
for ind,row in df.iterrows():
    df.loc[ind,"Study_Percentage"]=(row['10percentage']+row['12percentage']+row['collegeGPA'])/3
```

```python
print(df['Study_Percentage'])
```

```python
def corr_coeff(df,x,y):
# We need to calculate means of 2 columns 
    x_bar=df[x].mean()
    y_bar=df[y].mean()
# We need to calculate standard deviation of 2 columns
    x_sd=df[x].std()
    y_sd=df[y].std()
# We need to standardize the columns and multiply, then eventually sum all the products.
    total_prod=(((df[x]-x_bar)/x_sd)*((df[y]-y_bar)/y_sd)).sum()
    corr=total_prod/3997
    return corr
```

```python
corr_coeff(df,'Salary','Study_Percentage')
```

<h10>Question <b>4. Interpretation-</b>The correlation coefficient is 0.2021. That means there is a positive correlation between the variables 'Study_Percentage' and 'Salary'. In simple terms if the percentage is high, comaparably the salary package is high and viceversa. </h10>
