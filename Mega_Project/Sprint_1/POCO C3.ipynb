```python
# Scrapping the Web Page

Review_title = []
Review_text = []
Review_rating = []
User_city = []
Review_upvotes = []
Review_downvotes = []
Review_age = []


for i in range(1, 1116):
    URL='https://www.flipkart.com/poco-c3-arctic-blue-32-gb/product-reviews/itm815e140cbd2a0?pid=MOBFVQJ5PMVMSJHT&lid=LSTMOBFVQJ5PMVMSJHTHSBVUC&marketplace=FLIPKART&page={}'.format(i)
    page = requests.get(URL)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for x in soup.find_all('div', attrs={'class' : 'col _2wzgFH K0kLPL'}):

        Reviewtitle = x.find('p', attrs={'class' : "_2-N8zT"})
        if Reviewtitle is None:
            Review_title.append(np.NaN)
        else:
            Review_title.append(Reviewtitle.text)

        Reviewtext = x.find('div', attrs={'class' : 't-ZTKy'})
        if Reviewtext is None:
            Review_text.append(np.NaN)
        else:
            Review_text.append(Reviewtext.text)

        Reviewrating=soup.find('div', attrs={'class' : '_3LWZlK _1BLPMq'})
        if Reviewrating is None:
            Review_rating.append(np.NaN)
        else:
            Review_rating.append(Reviewrating.text)

        Usercity = x.find('p', attrs={'class' : "_2mcZGG"})
        if Usercity is None:
            User_city.append(np.NaN)
        else:
            User_city.append(Usercity.text)
        
        ReviewUpvotes = x.find('div', attrs={'class' : "_1LmwT9"})
        if ReviewUpvotes is None:
            Review_upvotes.append(np.NaN)
        else:
            Review_upvotes.append(ReviewUpvotes.text)
            
        ReviewDownvotes = x.find('div', attrs={'class' : "_1LmwT9 pkR4jH"})
        if ReviewDownvotes is None:
            Review_downvotes.append(np.NaN)
        else:
            Review_downvotes.append(ReviewDownvotes.text)
        
        ReviewAge = x.find('div', attrs={'class' : "row _3n8db9"})
        if ReviewAge is None:
            Review_age.append(np.NaN)
        else:
            Review_age.append(ReviewAge.text)  
```
