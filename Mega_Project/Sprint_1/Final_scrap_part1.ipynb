```python
import regex as re
```

```python
regex=r'[^,]*$'
df['User_city']=df['User_city'].apply(lambda x : re.compile(regex).findall(x))
```

```python
df['Review_age'].str.lower()
```

```python
regex1=r'\d+\s+[month|months|day|days]+'
regex2=r'[jan|january|feb|february|mar|march|apr|april|may|jun|june|july|aug|august|sep|september|oct|october|nov|november|dec|december]+\d\d\d\d'
regex3=r'today'
regex4=r'None'
for i in df['Review_age']:
    if 'month' in i:
        df['Review_age']=df['Review_age'].apply(lambda x : re.compile(regex1).findall(x))
    
    elif 'jan' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'january' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'feb' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'february' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'mar' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'march' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'apr' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'april' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'may' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'june' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'july' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'aug' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'august' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'sep' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'september' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'oct' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'october' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'nov' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'november' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'dec' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x))
    elif 'december' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex2).findall(x)) 
    elif 'today' in i:
        df['Review_age']=i.apply(lambda x : re.compile(regex3).findall(x)) 
    else:
        None
```
